```python
'''
Problem 1: Basic Line Plot
Scenario: You want to visualize a student's marks growth over 5 years.
Problem Statement:
Plot the years on the x-axis and marks on the y-axis.
Input: Years = [2018, 2019, 2020, 2021, 2022]
Marks = [65, 70, 72, 78, 85]
Tasks:

Use a red colored line.
Label x-axis as "Year" and y-axis as "Marks".
Give the plot a title: "Student Marks Growth".
'''

import matplotlib.pyplot as plt

years = [2018, 2019, 2020, 2021, 2022]
marks = [65, 70, 72, 78, 85]
plt.plot(years, marks, color='red')
plt.xlabel('Year')
plt.ylabel('Marks')
plt.title('Student Marks Growth')
plt.xticks(years)
plt.show()
```

```python
'''
Problem 2: Line Plot with Different Styles
Scenario: Company's profit over time.
Problem Statement:
Plot the years vs profits with a dashed line style.
Input: Years = [2016, 2017, 2018, 2019, 2020]
Profit = [100, 120, 150, 180, 200]
Tasks:

Use a green dashed line (linestyle='--').
Add title: "Company Profit Trend".
Show gridlines on the plot.
'''

import matplotlib.pyplot as plt
years = [2016, 2017, 2018, 2019, 2020]
profit = [100, 120, 150, 180, 200]

plt.plot(years, profit, color='green', linestyle='--')
plt.title('Company Profit Trend')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.grid()
plt.xticks(years)
plt.show()
```

```python
'''
Problem 3: Multiple Lines in a Single Plot
Scenario: Compare temperature in two cities over months.
Problem Statement:
Plot two lines in the same graph.
Input:
Months = [1, 2, 3, 4, 5, 6]
City1 Temp = [30, 32, 35, 37, 40, 42]
City2 Temp = [20, 22, 24, 27, 30, 33]
Tasks:

Use different colors for each city.
Add a legend.
Add title: "Temperature Comparison".
'''

import matplotlib.pyplot as plt
months = [1, 2, 3, 4, 5, 6]
city1_temp = [30, 32, 35, 37, 40, 42]
city2_temp = [20, 22, 24, 27, 30, 33]

plt.plot(months, city1_temp, color='red', label='City 1')
plt.plot(months, city2_temp, color='green', label='City 2')
plt.title('Temperature Comparison')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend()

plt.xticks(months)
plt.show()
```

```python
'''
Problem 4: Plot with Markers
Scenario: A runner's performance every day.
Problem Statement:
Plot distance run per day using markers.
Input:
Days = [1, 2, 3, 4, 5, 6, 7]
Distance = [2, 3, 5, 6, 5, 7, 8]
Tasks:

Use circle markers (marker='o').
Line color: blue.
Label x-axis and y-axis properly.
'''

import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5, 6, 7]
distance = [2, 3, 5, 6, 5, 7, 8]

plt.plot(days, distance, color='blue', marker='o')
plt.title('Runner Performance')
plt.xlabel('Days')
plt.ylabel('Distance (km)')
plt.xticks(days)
plt.show()
```

```python
'''
Problem 5: Line Color and Width
Scenario: Stock market value changes in a week.
Problem Statement:
Draw a plot with custom line width and color.
Input:
Days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
Stock Value = [100, 105, 102, 108, 110]
Tasks:

Line color: purple.
Line width: 3.
Title: "Stock Market Variation".
'''

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
stock_value = [100, 105, 102, 108, 110]

import matplotlib.pyplot as plt
plt.plot(days, stock_value, color='purple', linewidth=3)
plt.title('Stock Market Variation')
plt.show()
```

```python
'''
Problem 6: Line Plot with Different Markers
Scenario: Water consumption each month.
Problem Statement:
Create a line plot showing monthly water consumption.
Input:
Months = [1, 2, 3, 4, 5, 6]
Consumption = [300, 280, 290, 310, 320, 330]
Tasks:

Use square markers (marker='s').
Line color: brown.
Add gridlines.
'''

import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5, 6]
consumption = [300, 280, 290, 310, 320, 330]

plt.plot(months, consumption, color='brown', marker='s')
plt.title('Monthly Water Consumption')
plt.xlabel('Month')
plt.ylabel('Consumption (liters)')
plt.grid()
plt.xticks(months)
plt.show()
```

```python
'''
Problem 7: Line Plot with X-axis Rotation
Scenario: Sales numbers in different months.
Problem Statement:
Rotate x-axis labels for better visibility.
Input:
Months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
Sales = [200, 220, 250, 230, 240, 260]
Tasks:

Rotate x-axis labels by 45 degrees.
Title: "Monthly Sales".
Line color: orange.
'''

import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [200, 220, 250, 230, 240, 260]

plt.plot(months, sales, color='orange')
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(rotation=45)

plt.show()
```

```python
'''
Problem 8: Line Plot with Annotations
Scenario: Peak point in production.
Problem Statement:
Highlight the peak production value.
Input:
Days = [1, 2, 3, 4, 5]
Production = [400, 420, 450, 470, 500]
Tasks:

Mark the maximum point using annotate().
Line color: teal.
Title: "Production Over Days".
'''

import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
production = [400, 420, 450, 470, 500]

plt.plot(days, production, color='teal')
plt.title('Production Over Days')
plt.xlabel('Days')
plt.ylabel('Production')
plt.xticks(days)

# plt.annotate('Peak Production', xy=(5, 500), xytext=(3, 480), arrowprops=dict(facecolor='black', shrink=0.05))
max_value = max(production)
max_index = production.index(max_value)
plt.annotate('Peak Production', xy=(days[max_index], max_value), xytext=(days[max_index]-2, max_value-20), arrowprops=dict(arrowstyle='-|>'))
plt.show()
```

```python
'''
Problem 9: Plot Mathematical Functions
Scenario: Plotting simple functions.
Problem Statement:
Plot y = x² over a range.
Input:
x = range(-10, 11)
y = x²
Tasks:

Use a solid black line.
Label axes as "x" and "y".
Title: "Graph of y = x²".
'''

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 11, 1)
y = x ** 2
plt.plot(x, y, color='black')
plt.title('Graph of y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
'''
Problem 10: Line Plot with Custom Figure Size
Scenario: Displaying electricity usage.
Problem Statement:
Change the size of the figure.
Input:
Days = [1, 2, 3, 4, 5, 6, 7]
Usage = [30, 32, 31, 33, 35, 34, 36]
Tasks:

Set figure size to (10, 5).
Use magenta line color.
Title: "Electricity Usage Over a Week".
'''

import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
usage = [30, 32, 31, 33, 35, 34, 36]

plt.figure(figsize=(10, 5))
plt.plot(days, usage, color='magenta')
plt.title('Electricity Usage Over a Week')
plt.xlabel('Days')
plt.ylabel('Usage (kWh)')
plt.xticks(days)
plt.show()
```
