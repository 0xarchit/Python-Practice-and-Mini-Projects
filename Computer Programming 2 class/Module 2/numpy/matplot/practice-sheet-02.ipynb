```python
'''
Problem 1: Simple Line Plot
Scenario:
You are studying the growth of a plant over time.

Problem Statement:
Plot the height of a plant over 5 days.

Input Format:

Days: [1, 2, 3, 4, 5]
Height (cm): [2, 3, 5, 7, 11]
Output Format:

A line plot showing days on the x-axis and height on the y-axis.
Sample Output:

A simple upward curve showing plant growth.
'''

import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5]
height = [2, 3, 5, 7, 11]

plt.plot(days, height)
plt.title("Plant Growth Over 5 Days")
plt.xlabel("Days")
plt.ylabel("Height (cm)")
plt.show()
```

```python
'''
Problem 2: Scatter Plot of Students' Marks
Scenario:
You have students' marks in Math.

Problem Statement:
Plot marks using a scatter plot.

Input Format:

Students: [1, 2, 3, 4, 5]
Marks: [88, 92, 75, 85, 90]
Output Format:

Scatter plot with Students on x-axis and Marks on y-axis.
Sample Output:

Random dots showing variation of marks.
'''
import matplotlib.pyplot as plt

students = [1, 2, 3, 4, 5]
marks = [88, 92, 75, 85, 90]

plt.scatter(students, marks)
plt.title("Students' Marks in Math")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
'''
Problem 3: Bar Chart of Monthly Sales
Scenario:
You manage a store and track monthly sales.

Problem Statement:
Create a bar chart to represent sales.

Input Format:

Months: ['Jan', 'Feb', 'Mar', 'Apr']
Sales: [2500, 2700, 3000, 2800]
Output Format:

Bar chart showing sales.
Sample Output:

Bars of different heights representing sales.
'''

import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [2500, 2700, 3000, 2800]

plt.bar(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.show()
```

```python
'''
Problem 4: Multiple Line Plot
Scenario:
You are comparing temperatures of two cities over a week.

Problem Statement:
Plot temperatures of City A and City B on the same graph.

Input Format:

Days: [Mon, Tue, Wed, Thu, Fri]
City A Temp: [22, 24, 19, 23, 25]
City B Temp: [18, 20, 17, 21, 22]
Output Format:

Two lines with legends (City A and City B).
Sample Output:

Two distinct colored lines crossing or following each other.
'''

import matplotlib.pyplot as plt
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
city_a = [22, 24, 19, 23, 25]
city_b = [18, 20, 17, 21, 22]

plt.plot(days, city_a, label='City A', marker='o')
plt.plot(days, city_b, label='City B', marker='o')
plt.title("Temperature Comparison")
plt.show()
```

```python
'''
Problem 5: Pie Chart of Favorite Fruits
Scenario:
You surveyed students about favorite fruits.

Problem Statement:
Create a pie chart.

Input Format:

Fruits: ['Apple', 'Banana', 'Cherry', 'Mango']
Votes: [40, 25, 20, 15]
Output Format:

A pie chart showing percentages.
Sample Output:

Sectors of different sizes for each fruit.
'''

import matplotlib.pyplot as plt
fruits = ['Apple', 'Banana', 'Cherry', 'Mango']
votes = [40, 25, 20, 15]
plt.pie(votes, labels=fruits, autopct='%0.0f%%')
plt.title("Favorite Fruits")
plt.show()
```

```python
'''
Problem 6: Histogram of Ages
Scenario:
You collected ages of people attending a workshop.

Problem Statement:
Create a histogram.

Input Format:

Ages: [22, 25, 30, 22, 29, 40, 35, 25, 28, 30, 32, 40, 22]
Output Format:

Histogram showing the frequency of each age range.
Sample Output:

Bars showing age distribution.
'''

import matplotlib.pyplot as plt
ages = [22, 25, 30, 22, 29, 40, 35, 25, 28, 30, 32, 40, 22]
plt.hist(ages, bins=5, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.xticks(list(set(ages)))
plt.show()
```

```python
'''
Problem 7: Subplots (Marks vs Attendance)
Scenario:
You want to analyze if attendance impacts marks.

Problem Statement:
Create subplots: one for attendance, one for marks.

Input Format:

Students: [1, 2, 3, 4, 5]
Attendance (%): [90, 85, 80, 95, 70]
Marks: [88, 75, 70, 95, 60]
Output Format:

Two side-by-side plots: one for attendance, one for marks.
Sample Output:

First plot: bar plot for attendance, second: bar plot for marks.
'''
import matplotlib.pyplot as plt
students = [1, 2, 3, 4, 5]
att = [90, 85, 80, 95, 70]
marks = [88, 75, 70, 95, 60]

plt.subplot(1, 2, 1)
plt.bar(students, att, color='blue')
plt.title("Attendance")
plt.xlabel("Students")
plt.ylabel("Attendance (%)")

plt.subplot(1, 2, 2)
plt.bar(students, marks, color='green')
plt.title("Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
'''
Problem 8: Filled Area Between Two Lines
Scenario:
You track the high and low temperatures of a city.

Problem Statement:
Plot and shade the area between high and low temperature curves.

Input Format:

Days: [1, 2, 3, 4, 5, 6, 7]
High Temp: [30, 32, 33, 35, 36, 34, 33]
Low Temp: [20, 21, 19, 22, 23, 21, 20]
Output Format:

Fill the area between high and low temperature lines.
Sample Output:

Shaded area showing the temperature range each day.
'''
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7]
high_temp = [30, 32, 33, 35, 36, 34, 33]
low_temp = [20, 21, 19, 22, 23, 21, 20]

plt.plot(days, high_temp, label='High Temp', color='red')
plt.plot(days, low_temp, label='Low Temp', color='blue')
plt.fill_between(days, high_temp, low_temp, color='lightgray')
plt.title("Temperature Range Over a Week")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```

```python
'''
Problem 9: Heatmap of Student Scores
Scenario:
You have scores of students in multiple subjects.

Problem Statement:
Create a heatmap.

Input Format:

2D scores array:
[
  [80, 85, 90],
  [70, 75, 80],
  [60, 65, 70],
  [90, 95, 85]
]
Subjects: ['Math', 'Science', 'English']
Output Format:

Heatmap where higher scores are brighter.
Sample Output:

Colored squares with intensity showing score levels.
'''
import matplotlib.pyplot as plt
import numpy as np

scores = np.array([[80, 85, 90], 
                   [70, 75, 80],
                   [60, 65, 70],
                   [90, 95, 85]])
subjects = ['Math', 'Science', 'English']

plt.imshow(scores, cmap='hot', interpolation='nearest')
plt.colorbar(label='Scores')
plt.xticks(np.arange(len(subjects)), subjects)
plt.yticks(np.arange(len(scores)), ['Student 1', 'Student 2', 'Student 3', 'Student 4'])
plt.title("Student Scores Heatmap")
plt.xlabel("Subjects")
plt.ylabel("Students")
plt.show()
```

```python
'''
Problem 10: Contour Plot of a Mathematical Function
Scenario:
You want to visualize a 3D function on 2D.

Problem Statement:
Create a contour plot for ( Z = X^2 + Y^2 ).

Input Format:

X and Y range: -5 to 5
Output Format:

Contour plot showing lines of equal value.
Sample Output:

Circular contours centered at (0,0).
'''

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

plt.contour(X, Y, Z, levels=20)
plt.colorbar(label='Z = X^2 + Y^2')
plt.title("Contour Plot of Z = X^2 + Y^2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
