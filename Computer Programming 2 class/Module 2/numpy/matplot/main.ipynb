```python
import matplotlib.pyplot as plt

marks1 = [40, 60, 35, 78, 20, 50]
marks2 = [50, 70, 45, 88, 30, 60]
# plt.plot(marks, marker='o', linestyle='--', color='b')
plt.plot(range(1, 7), marks1, marker='o', linestyle='--', color='b', label='Student 1')
plt.plot(range(1, 7), marks2, marker='o', linestyle='-.', color='r', label='Student 2')
plt.title('Marks of Students')
plt.xlabel('Subject Number')
plt.ylabel('Marks')
plt.grid()

for i, mark in enumerate(marks1, 1): # Start enumeration from 1 return tuple of counter and value
    if mark < 40:
        plt.text(i , mark + 1, str(mark), ha='center', va='bottom', color='red')
    elif mark < 60:
        plt.text(i , mark + 1, str(mark), ha='center', va='bottom', color='orange')
    else:
        plt.text(i , mark + 1, str(mark), ha='center', va='bottom', color='green')

for i, mark in enumerate(marks2, 1):
    if mark < 40:
        plt.text(i , mark + 1, str(mark), ha='center', va='bottom', color='red')
    elif mark < 60:
        plt.text(i , mark + 1, str(mark), ha='center', va='bottom', color='orange')
    else:
        plt.text(i , mark + 1, str(mark), ha='center', va='bottom', color='green')

plt.axhline(y=40, color='r', linestyle='--', label='Pass Mark')
plt.axhline(y=60, color='g', linestyle='--', label='Distinction Mark')
plt.legend()

plt.xticks(range(1, 7), ['Web', 'Pyton', 'English', 'Physics', 'Math', 'Design Thinking'])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x over a full cycle from 0 to 2*pi
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.axhline(0, color='red', lw=0.5, ls='--')
plt.axvline(0, color='red', lw=0.5, ls='--')
plt.fill_between(x, y_sin, color='blue', alpha=0.1)
plt.fill_between(x, y_cos, color='red', alpha=0.1)

plt.title('Sine and Cosine')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
# Create the x values and compute the functions
x = np.linspace(-2*np.pi, 2*np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Mask out extreme tan values for a cleaner plot
threshold = 10
y_tan_clean = np.copy(y_tan)
y_tan_clean[np.abs(y_tan_clean) > threshold] = np.nan

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, y_tan_clean, label='tan(x)', color='green')
plt.title('Tangent Function')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.ylim(-threshold, threshold)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
